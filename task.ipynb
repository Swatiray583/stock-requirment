```python
import pandas as pd
```

```python
inventory_df=pd.read_csv("C:/Users/ankit/Downloads/Inventory.csv")
sales_df=pd.read_excel("C:/Users/ankit/Downloads/Sales.xlsx")
```

```python
aggregated_inventory = inventory_df.groupby('Sku Code').agg({
    'SKU Name': 'first',
    'Total': 'sum',
    'Free Qty': 'sum'
}).reset_index()
```

```python
#aggregated_inventory['Current_Stock'] = aggregated_inventory['Total'] - aggregated_inventory['Free Qty']
```

```python
aggregated_sales = sales_df['Sku Code'].value_counts().reset_index()
aggregated_sales.columns = ['Sku Code', 'Quantity_Sold']
```

```python
merged_df = pd.merge(aggregated_inventory, aggregated_sales, on='Sku Code', how='left').fillna(0)
merged_df['Quantity_Sold'] = merged_df['Quantity_Sold'].astype(int)
```

```python
merged_df['Required_Stock'] = merged_df['Quantity_Sold'] - merged_df['Free Qty']
merged_df['Restock_Needed'] = merged_df['Required_Stock'].apply(lambda x: 'Yes' if x > 0 else 'No')
```

```python
output_file = 'Required_Stock.xlsx'
merged_df.to_excel(output_file, index=False)

print(f"Required stock data has been saved to {output_file}.")
```
